### F1. Shopping List Manager (1D list + functions) (Suggested)
Write a program that maintains a shopping list (a 1D list of strings).

**Required functions:**
- `add_item(items, item)`
- `remove_item(items, item)` (remove by value if present)
- `print_items(items)` (numbered list)
- `save_items(items, filename)` (one item per line)
- `load_items(filename)` (returns a list)

**Program behavior:**
Provide a simple text menu in a loop:
1) add  2) remove  3) view  4) save  5) load  6) quit
The list should update as the user makes choices.


````python
# TODO: Implement F1 here

def add_item(items, item):
  items.append(item)
  return items

def remove_item(items, item):
  if item in items:
    items.remove(item)
  return items

def print_items(items):
  if not items:
    print("The shopping list is empty.")

  for i, item in enumerate(items, start=1):
      print(f"{i}. {item}")

def save_items(items, filename):
  with open(filename, 'w') as f:
    for item in items:
      f.write(item + '\n') # Added newline character

def load_items(filename): # Corrected function name
  items = []
  try:
      with open(filename, "r") as file:
          for line in file:
              items.append(line.strip())
      return items
  except FileNotFoundError:
      print("File not found.")
      return []

def main():

  items = []

  while True:
    print('\nShopping List Menu:')
    print('--------------------')
    print('1. Add Item\n2. Remove Item\n3. View List\n4. Save List')
    print('5. Load List\n6. Quit\n')

    user_input = input('Please insert a number between 1 to 6 to edit, view, and save your list. ')

    if user_input == "1":
            item = input("Enter item to add: ")
            items = add_item(items, item)

    elif user_input == "2":
        item = input("Enter item to remove: ")
        items = remove_item(items, item)

    elif user_input == "3":
        print_items(items)

    elif user_input == "4":
        filename = input("Enter filename to save to (Name the file): ")
        save_items(items, filename)

    elif user_input == "5":
        filename = input("Enter filename to load from: ")
        items = load_items(filename)

    elif user_input == "6":
        print('\nHere is your finished list:')
        print_items(items)

        break

    else:
      print('Invalid input. Please input the given choices.')



main()
````

### F2. Grade Statistics (1D list of numbers + file input)
You will read a file of grades (one number per line) into a 1D list, then compute statistics.

**Requirements:**
- Read `grades.txt` into a list `grades`
- Compute and print: count, min, max, average
- Create a new list `curved` that adds +5 to each grade (cap at 100)
- Write `curved` to `curved_grades.txt` (one per line)

**If you don't have a `grades.txt` file, create one in code first** (write 8–12 sample grades).


````python
# TODO: Implement F2 here
def grades_calc():

  grades = [66, 96, 86, 72, 83, 91, 65, 79, 98]

  with open('grades.txt', 'w') as f:
    for grade in grades:
      f.write(str(grades) + '\n')

  with open('grades.txt', 'r') as f:
    for grade in range(len(grades)):
      grades[grade] = int(grades[grade])

    print('Original Grades:', grades)

    min_grade = min(grades)
    max_grade = max(grades)
    total = len(grades)
    avg = sum(grades)/total

    print('Minimum:', min_grade)
    print('Maximum:', max_grade)
    print('Total:', total)
    print(f'Average:{avg:.2f}')

  with open('cureved_grades.txt', 'w') as f:
    curved = []
    for grade in grades:
        if grade <= 95:
           grade += 5
           curved.append(grade)
        else:
          grade = 100
          curved.append(grade)

    for grade in grades:
      f.write(str(curved) + '\n')

  with open('cureved_grades.txt', 'r') as f:
    for grade in range(len(grades)):
      grades[grade] = int(grades[grade])

    print('Curved Grades:', curved)

grades_calc()
````

### F4. Seating Chart (nested list) (2D list + loops + functions)
Create a seating chart using a nested list (list of rows).

**Requirements:**
- Build a 4×5 chart filled with the string `"EMPTY"`
- Write a function `assign_seat(chart, row, col, name)` that places a name
- Write a function `print_chart(chart)` that prints each row neatly
- Assign at least 6 names to different seats
- Count how many seats are still `"EMPTY"` and print that number

**Optional extension:** Write the seating chart to a file (one row per line).


````python
# TODO: Implement F4 here
#import random

row = 4
col = 5

count = 0

chart = [['EMPTY' for _ in range(col)] for _ in range(row)]

def assign_seat(chart, row, col, name):
  chart[row][col] = name

def print_chart(chart):
  for row in chart:
    print(row)

assign_seat(chart, 3, 2, 'Sara')
assign_seat(chart, 1, 3, 'Leah')
assign_seat(chart, 0, 1, 'Aviva')
assign_seat(chart, 3, 4, 'Amy')
assign_seat(chart, 1, 0, 'Frieda')
assign_seat(chart, 2, 2, 'Esther')
assign_seat(chart, 3, 0, 'Michelle')
assign_seat(chart, 3, 3, 'Kayla')
assign_seat(chart, 2, 1, 'Ruchama')
assign_seat(chart, 1, 2, 'Deborah')

print_chart(chart)

for row in chart:
  for seat in row:
    if seat == 'EMPTY':
      count += 1

print('\nEmpty Seat Count: ', count)

with open('seating_chart.txt', 'w') as f:
    for row in chart:
        f.write(str(row) + '\n')

````
